```python
# -- Heroes of Pymoli Challenge --

# Import Dependencies
import pandas as pd

# CSV file
file = 'Resources/purchase_data.csv'

# Read CSV file and store in Pandas data frame
purchase_data_df = pd.read_csv(file)
```

```python
# Grab all unique players IDs and display total players
unique_players_df = purchase_data_df.drop_duplicates(subset="SN", keep="last")

total_players = len(unique_players_df)
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
# Calculate 4 different stats from purchase data
unique_items = len(purchase_data_df["Item Name"].unique()) 
average_item_price = purchase_data_df["Price"].mean() 
total_purchases = len(purchase_data_df)
total_revenue = purchase_data_df["Price"].sum()

# Insert summary stats to data frame
summary_stats_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                 "Average Item Price": [average_item_price],
                                 "Number Of Purchases": [total_purchases],
                                 "Total Revenue": [total_revenue]})

# Map to format Item Price and Revenue
summary_stats_df["Average Item Price"] = summary_stats_df["Average Item Price"].map("${:.2f}".format)
summary_stats_df["Total Revenue"] = summary_stats_df["Total Revenue"].map("${:.2f}".format)
summary_stats_df
```

```python
# Group by gender 
gender_group = unique_players_df.groupby(["Gender"])
gender_demo_df = gender_group.count()
gender_demo_df = gender_demo_df[["Purchase ID"]]
gender_demo_df = gender_demo_df.rename(
    columns={"Purchase ID": "Total Count"})
# Get percentage from total players
percentage = (gender_demo_df["Total Count"] / total_players) * 100
gender_demo_df["Percentage of Players"] = percentage
gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].map("{:.2f}%".format)
gender_demo_df = gender_demo_df.sort_values("Total Count", ascending = False)
gender_demo_df.head()
```

```python
gender_sales_df = purchase_data_df[["Gender", "Price"]]
gender_sales = gender_sales_df.groupby(["Gender"])
purchase_count_df = gender_sales.count() 
gender_stats_df = purchase_count_df.rename(
    columns={"Price": "Purchase Count"})
purchase_average = gender_sales.mean()
purchase_total = gender_sales.sum()
gender_stats_df["Average Purchase Price"] = purchase_average
gender_stats_df["Total Purchase Value"] = purchase_total
gender_stats_df = gender_stats_df.sort_values("Purchase Count", ascending = False)

player_count = gender_demo_df["Total Count"]
person_average = gender_stats_df["Total Purchase Value"] / player_count
gender_stats_df["Avg. Total Purchase Per Person"] = person_average

gender_stats_df["Average Purchase Price"] = gender_stats_df["Average Purchase Price"].map("${:.2f}".format)
gender_stats_df["Total Purchase Value"] = gender_stats_df["Total Purchase Value"].map("${:.2f}".format)
gender_stats_df["Avg. Total Purchase Per Person"] = gender_stats_df["Avg. Total Purchase Per Person"].map("${:.2f}".format)


gender_stats_df
```

```python
# Define bins and labels for them
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 130]
group_names = ["<10", "10-14", "15-19","20-24", "25-29", "30-34", "35-39", "40+"]

# Copy unique player list and bin by age range
ages_df = unique_players_df.copy() 
ages_df["Age Groups"] = pd.cut(ages_df["Age"], age_bins, labels=group_names)

# Count players in age groups and perform calculation on percentage
ages_grouped = ages_df.groupby(["Age Groups"])
ages_grouped_df = ages_grouped.count()
ages_grouped_df = ages_grouped_df.rename(
    columns={"Purchase ID": "Total Count"})
ages_grouped_df = ages_grouped_df[["Total Count"]]
ages_percentage = (ages_grouped_df["Total Count"] / total_players) * 100
ages_grouped_df["Percentage of Players"] = ages_percentage

# Format for percentage
ages_grouped_df["Percentage of Players"] = ages_grouped_df["Percentage of Players"].map("{:.2f}%".format)

ages_grouped_df
```

```python
# Bin to age groups
age_purchase_df = purchase_data_df.copy()
age_purchase_df["Age Ranges"] = pd.cut(age_purchase_df["Age"], age_bins, labels=group_names)
age_purchase_df

# Group by the age ranges
age_purchases = age_purchase_df.groupby(["Age Ranges"])
purchase_count_df = age_purchases.count()
purchase_stats_df = purchase_count_df.rename(
    columns={"Price": "Purchase Count"})

# Calculate stats and put into dataframe
age_average = age_purchases.mean()
age_total = age_purchases.sum()
purchase_stats_df["Average Purchase Price"] = age_average["Price"]
purchase_stats_df["Total Purchase Value"] = age_total["Price"]
player_average = purchase_stats_df["Total Purchase Value"] / ages_grouped_df["Total Count"]
purchase_stats_df["Avg. Total Purchase Per Person"] = player_average
purchase_stats_df = purchase_stats_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg. Total Purchase Per Person"]]

# Format columns and clean up dataframe
purchase_stats_df["Average Purchase Price"] = purchase_stats_df["Average Purchase Price"].map("${:.2f}".format)
purchase_stats_df["Total Purchase Value"] = purchase_stats_df["Total Purchase Value"].map("${:.2f}".format)
purchase_stats_df["Avg. Total Purchase Per Person"] = purchase_stats_df["Avg. Total Purchase Per Person"].map("${:.2f}".format)

purchase_stats_df
```

```python
purchase_group = purchase_data_df.groupby(["SN"])
top_spender_df = purchase_group.count()

top_spender_df  = top_spender_df[["Purchase ID"]]
top_spender_df  = top_spender_df.rename(
    columns={"Purchase ID": "Purchase Count"})
purchase_count = top_spender_df["Purchase Count"]
purchase_total = purchase_group.sum()
purchase_average = purchase_total["Price"] / purchase_count
purchase_average
top_spender_df["Average Purchase Price"] = purchase_average
top_spender_df["Total Purchase Value"] = purchase_total["Price"]
top_spender_df = top_spender_df.sort_values("Total Purchase Value", ascending = False)

top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)

top_spender_df.head(5)
```

```python
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
id_group = items_df.groupby(["Item ID", "Item Name"])
items_summary_df = id_group.count()
item_total = id_group.sum()
item_price = id_group.sum() /id_group.count()
items_summary_df["Item Price"] = item_price
items_summary_df["Total Purchase Value"] = item_total
items_summary_df = items_summary_df.sort_values("Price", ascending = False)
items_summary_df  = items_summary_df.rename(
    columns={"Price": "Purchase Count"})

items_summary_df["Item Price"] = items_summary_df["Item Price"].map("${:.2f}".format)
items_summary_df["Total Purchase Value"] = items_summary_df["Total Purchase Value"].map("${:.2f}".format)

items_summary_df.head()
```

```python
items_value_df = items_summary_df.copy()
items_value_df["Total Purchase Value"] = items_value_df["Total Purchase Value"].str.replace("$", "")
items_value_df["Total Purchase Value"] = items_value_df["Total Purchase Value"].astype(float)
items_value_df = items_value_df.sort_values("Total Purchase Value", ascending = False)
items_value_df["Total Purchase Value"] = items_value_df["Total Purchase Value"].map("${:.2f}".format)
items_value_df.head()
```
